```python
import numpy as np
from kmeansCluster import kmeansCluster
from gmmCluster import gmmCluster

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

# for auto-reloading external modules
# see http://stackoverflow.com/questions/1907993/autoreload-of-modules-in-ipython
%load_ext autoreload
%autoreload 2
```

```python
def loadData():
    Dataset1 = np.loadtxt("./data_sets_clustering/humanData.txt")
    Dataset2 = np.loadtxt("./data_sets_clustering/audioData.txt")

    X1=Dataset1[:,:2]
    Y1=Dataset1[:,2]
    X2=Dataset2[:,:13]
    Y2=Dataset2[:,13]
    return X1,Y1,X2,Y2

X1,Y1,X2,Y2 = loadData()
```

```python
k=2
kmeans = kmeansCluster(k,1000)
kmeans.run(X1)
clusters = kmeans.clusters
labels = kmeans.xlabel

plt.figure()
plt.scatter(X1[:,0],X1[:,1],c=labels)
plt.title('Kmeans clustering when k = %d' %k)
plt.show()
```

```python
# k_knee function

def k_knee(X,k_range):
    Obj_fun=[]
    for k in k_range:
        obj=0
        kmeans = kmeansCluster(k,1000)
        kmeans.run(X1)
        clusters = kmeans.clusters
        labels = kmeans.xlabel
        for i in range(k):
            obj=obj+np.sum((X[labels==i]-clusters[i])**2)
        Obj_fun.append(obj)

    plt.figure()
    plt.plot(k_range,Obj_fun)
    plt.title("k_knee")
    plt.xlabel("k")
    plt.ylabel("Objective function")
    plt.show()
    return Obj_fun


k_range = [1,2,3,4,5,6]
Obj_fun = k_knee(X1,k_range)
```

```python
# purity when k = 2
import collections
def return_max(count_dict):
    max_value= - 10**99
    for (k,v) in collections.Counter(count_dict).items():
        if v > max_value:
            max_value = v
            max_key = k
    return max_key, max_value


# Q5
def purity(labels,trueLabels):
    p = []
    for (k,v) in collections.Counter(labels).items():
        temp = trueLabels[labels==k]
        max_key, max_value = return_max(temp)
        p.append((k,max_value*1.0/v))
    return p

purityMetric = purity(labels,Y1)
print "(cluster , purityMetric) for Kmeans is "+str(purityMetric)
```

```python
# gmmCluster

k = 2
gmm = gmmCluster(k = k, covType = "diag", maxIter = 1000)
gmm.run(X1)
mu_diag = gmm.mu
labels_diag = gmm.labels
plt.figure()
plt.scatter(X1[:,0],X1[:,1],c=np.array(labels_diag))
plt.title('GMM clustering when k = '+ str(k)+" Sigma is diagonal")
plt.show()
```

```python
# try gmmCluster on Dataset1 when using 'diag' or 'full' covariance matrix

# diag
k = 2
gmm = gmmCluster(k = k, covType = "diag", maxIter = 1000)
gmm.run(X1)
labels_1diag = gmm.labels

plt.figure()
plt.scatter(X1[:,0],X1[:,1],c=np.array(labels_1diag))
plt.title('GMM clustering when k = '+ str(k)+" Sigma is diagonal")
plt.show()

# full 
k = 2
gmm = gmmCluster(k = k, covType = "full", maxIter = 1000)
gmm.run(X1)
labels_1full = gmm.labels
plt.figure()
plt.scatter(X1[:,0],X1[:,1],c=np.array(labels_1full))
plt.title('GMM clustering when k = '+ str(k)+" Sigma is full")
plt.show()
```

```python
# try gmmCluster on Dataset2 when using 'diag' or 'full' covariance matrix

# diag

k = 2
gmm = gmmCluster(k = k, covType = "diag", maxIter = 1000)
gmm.run(X2)
labels_2diag = gmm.labels
plt.figure()
plt.scatter(X2[:,0],X2[:,1],c=np.array(labels_2diag))
plt.title('GMM clustering when k = '+ str(k)+" and Sigma is 'diag'")
plt.show()

# full
k = 2
gmm = gmmCluster(k = k, covType = "full", maxIter = 1000)
gmm.run(X2)
labels_2full = gmm.labels
plt.figure()
plt.scatter(X2[:,0],X2[:,1],c=np.array(labels_2full))
plt.title('GMM clustering when k = '+ str(k)+" and Sigma is 'full'")
plt.show()
```

```python
# calculate purity in the aboves cases

purityMetric_1diag = purity(labels_1diag,Y1)
print("(cluster , purityMetric) for GMM on dataset1 with diagSigma is "+str(purityMetric_1diag))

purityMetric_1full = purity(labels_1full,Y1)
print("(cluster , purityMetric) for GMM on dataset1 with fullSigma is "+str(purityMetric_1full))

purityMetric_2diag = purity(labels_2diag,Y2)
print("(cluster , purityMetric) for GMM on with diagSigma dataset2 is "+str(purityMetric_2diag))

purityMetric_2full = purity(labels_2full,Y2)
print("(cluster , purityMetric) for GMM on with fullSigma dataset2 is "+str(purityMetric_2full))
```
